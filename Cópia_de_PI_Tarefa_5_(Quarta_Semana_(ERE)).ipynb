#2. Filtragem

Considere os seguintes filtros:

*   Filtro da Média
*   Filtro Gaussiano
*   Filtro da Mediana
*   Filtro Sobel
*   Filtro Laplaciano

Crie uma função que receba uma imagem e o nome do filtro. Essa função deve retornar a imagem com o filtro aplicado. 

A seguir, apresente todas as imagens com os filtros aplicados em um grid. Use os subplots conforme apresentados no vídeo abaixo:
<a href="https://www.youtube.com/watch?v=U35qns3LvvI"><img src="https://img.youtube.com/vi/U35qns3LvvI/mqdefault.jpg"/></a>

Explique com suas palavras no áudio qual foi o efeito e quando cada filtro deve ser aplicado.

Referências:
<a href="https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html">Link 1</a> e <a href="https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_gradients/py_gradients.html">Link 2</a>.

```
#Insira seu código aqui
import numpy as np
import cv2

def filtroMedia(img):
  img = cv2.blur(img,(7,7))
  return img

def filtroGaussiano(img):
  img = cv2.GaussianBlur(img,(7,7),0)
  return img

def filtroMediana(img):
  img = cv2.medianBlur(img, 5)
  return img

def filtroSobel(img):
  img = cv2.Sobel(img, cv2.CV_8U, 1, 1, 5)
  return img

def filtroLaplaciano(img):
  img = cv2.Laplacian(img, cv2.CV_8U)
  return img

def mostraMultiplasImg(vetorImg, vetorTitulo, x , y):
  fig , axis =  plt.subplots(y, x)
  xId, yId , tituloId = 0, 0, 0
  for img in vetorImg:
      imgMPLIB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      axis[yId, xId].set_title(vetorTitulo[tituloId])
      axis[yId, xId].imshow(imgMPLIB)
           
      tituloId += 1
      xId += 1
      if xId == x:
        
        xId = 0
        yId += 1

      fig.tight_layout(pad = 0.5)
      
        
  plt.show()
  
def plotImagens():
  imgFace = cv2.imread("Facebook.jpg")
  imgFiltroMedia =  filtroMedia(imgFace)
  imgFiltroGuassiano = filtroGaussiano(imgFace)
  imgFiltroMediana = filtroMediana(imgFace)
  imgFiltroSobel = filtroSobel(imgFace)
  imgFiltroLaplace = filtroLaplaciano(imgFace)

  vetorImg = [imgFace, imgFiltroMedia, imgFiltroGuassiano, imgFiltroMediana, imgFiltroSobel, imgFiltroLaplace]
  vetorTitulo = ['Img Original', 'Img Filtro Media', 'Img Filtro Gaussiano', 'Img Filtro Mediana', 'Img Filtro Sobel', 'Img Filtro Laplace']
  mostraMultiplasImg(vetorImg, vetorTitulo, 3, 2)

from matplotlib import pyplot as plt

plotImagens()





```
